##  Morphological Filtering
1. Implementing Dilation and Erosion:

● Write code to perform dilation and erosion operations on a noisy grayscale image using Python.

● Experiment with different structuring elements (e.g., square, disk, diamond) and compare how they affect the output of dilation and erosion

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage import morphology
```

```python
# Load the color image
image_path = '/image.webp'  # Change this to your image file
color_image = cv2.imread(image_path)

# Convert to grayscale
gray_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
# Add random noise
noise = np.random.randint(0, 50, gray_image.shape, dtype=np.uint8)  # Adjust noise level as needed
noisy_image = cv2.add(gray_image, noise)

# Display the noisy image
plt.imshow(noisy_image, cmap='gray')
plt.title('Noisy Grayscale Image')
plt.axis('off')
plt.show()
```

```python
def dilate_image(image, selem):
    """Perform dilation on the image using the specified structuring element."""
    return morphology.dilation(image, selem)

def erode_image(image, selem):
    """Perform erosion on the image using the specified structuring element."""
    return morphology.erosion(image, selem)
```

```python
# Define structuring elements
square_selem = morphology.square(5)  # Square structuring element
disk_selem = morphology.disk(5)      # Disk structuring element
diamond_selem = morphology.diamond(5)  # Diamond structuring element

# Perform dilation and erosion with different structuring elements
dilated_square = dilate_image(noisy_image, square_selem)
dilated_disk = dilate_image(noisy_image, disk_selem)
dilated_diamond = dilate_image(noisy_image, diamond_selem)

eroded_square = erode_image(noisy_image, square_selem)
eroded_disk = erode_image(noisy_image, disk_selem)
eroded_diamond = erode_image(noisy_image, diamond_selem)

# Display results
fig, axes = plt.subplots(3, 2, figsize=(10, 10))
axes[0, 0].imshow(dilated_square, cmap='gray')
axes[0, 0].set_title('Dilation with Square SE')
axes[0, 0].axis('off')

axes[0, 1].imshow(eroded_square, cmap='gray')
axes[0, 1].set_title('Erosion with Square SE')
axes[0, 1].axis('off')

axes[1, 0].imshow(dilated_disk, cmap='gray')
axes[1, 0].set_title('Dilation with Disk SE')
axes[1, 0].axis('off')

axes[1, 1].imshow(eroded_disk, cmap='gray')
axes[1, 1].set_title('Erosion with Disk SE')
axes[1, 1].axis('off')

axes[2, 0].imshow(dilated_diamond, cmap='gray')
axes[2, 0].set_title('Dilation with Diamond SE')
axes[2, 0].axis('off')

axes[2, 1].imshow(eroded_diamond, cmap='gray')
axes[2, 1].set_title('Erosion with Diamond SE')
axes[2, 1].axis('off')

plt.tight_layout()
plt.show()
```


# **2. Opening Operation:**
● Implement the opening operation using OpenCV and logic.

● Apply the opening operation on a grayscale image using different structuring
elements such as rectangles, circles, and crosses.

● Compare the results obtained with different structuring elements

```python
# Create a simple binary image
image_path = '/image.webp'  # Change this to your image file
color_image = cv2.imread(image_path)

# Convert to grayscale
gray_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)
image = np.zeros((100, 100), dtype=np.uint8)
cv2.rectangle(image, (30, 30), (70, 70), 255, -1)  # Add a white square

# Add random noise
noise = np.random.randint(0, 50, image.shape, dtype=np.uint8)  # Adjust noise level as needed
noisy_image = cv2.add(image, noise)

# Display the noisy image
plt.imshow(noisy_image, cmap='gray')
plt.title('Noisy Grayscale Image')
plt.axis('off')
plt.show()
```

```python
def dilate_image(image, selem):
    return morphology.dilation(image, selem)

def erode_image(image, selem):
    return morphology.erosion(image, selem)
```

```python
square_selem = morphology.square(5)  # Square structuring element
disk_selem = morphology.disk(5)      # Disk structuring element
diamond_selem = morphology.diamond(5)  # Diamond structuring element

# Perform dilation and erosion with different structuring elements
dilated_square = dilate_image(noisy_image, square_selem)
dilated_disk = dilate_image(noisy_image, disk_selem)
dilated_diamond = dilate_image(noisy_image, diamond_selem)

eroded_square = erode_image(noisy_image, square_selem)
eroded_disk = erode_image(noisy_image, disk_selem)
eroded_diamond = erode_image(noisy_image, diamond_selem)

# Display results
fig, axes = plt.subplots(3, 2, figsize=(10, 10))
axes[0, 0].imshow(dilated_square, cmap='gray')
axes[0, 0].set_title('Dilation with Square SE')
axes[0, 0].axis('off')

axes[0, 1].imshow(eroded_square, cmap='gray')
axes[0, 1].set_title('Erosion with Square SE')
axes[0, 1].axis('off')

axes[1, 0].imshow(dilated_disk, cmap='gray')
axes[1, 0].set_title('Dilation with Disk SE')
axes[1, 0].axis('off')

axes[1, 1].imshow(eroded_disk, cmap='gray')
axes[1, 1].set_title('Erosion with Disk SE')
axes[1, 1].axis('off')

axes[2, 0].imshow(dilated_diamond, cmap='gray')
axes[2, 0].set_title('Dilation with Diamond SE')
axes[2, 0].axis('off')

axes[2, 1].imshow(eroded_diamond, cmap='gray')
axes[2, 1].set_title('Erosion with Diamond SE')
axes[2, 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
def opening_operation(image, selem):
    eroded = erode_image(image, selem)
    opened = dilate_image(eroded, selem)
    return opened
```

```python
rectangle_selem = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
circle_selem = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))  # Circular structuring element
cross_selem = cv2.getStructuringElement(cv2.MORPH_CROSS, (5, 5))    # Cross-shaped structuring element

# Apply the opening operation with different structuring elements
opened_rectangle = opening_operation(noisy_image, rectangle_selem)
opened_circle = opening_operation(noisy_image, circle_selem)
opened_cross = opening_operation(noisy_image, cross_selem)

# Display results of the opening operation
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
axes[0, 0].imshow(opened_rectangle, cmap='gray')
axes[0, 0].set_title('Opening with Rectangle SE')
axes[0, 0].axis('off')

axes[0, 1].imshow(opened_circle, cmap='gray')
axes[0, 1].set_title('Opening with Circle SE')
axes[0, 1].axis('off')

axes[1, 0].imshow(opened_cross, cmap='gray')
axes[1, 0].set_title('Opening with Cross SE')
axes[1, 0].axis('off')

plt.tight_layout()
plt.show()
```

# **3 Closing Operation:**
● Implement the closing operation using OpenCV and logic.

● Apply the closing operation on a grayscale image using different structuring
elements such as rectangles, circles, and crosses.

● Compare the results obtained with different structuring elements

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
image = cv2.imread('/image.webp', cv2.IMREAD_GRAYSCALE)

# Define the structuring elements for closing
rectangle_selem = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
circle_selem = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))  # Circular structuring element
cross_selem = cv2.getStructuringElement(cv2.MORPH_CROSS, (5, 5))    # Cross-shaped structuring element

# Apply the closing operation with different structuring elements
closed_rectangle = cv2.morphologyEx(image, cv2.MORPH_CLOSE, rectangle_selem)
closed_circle = cv2.morphologyEx(image, cv2.MORPH_CLOSE, circle_selem)
closed_cross = cv2.morphologyEx(image, cv2.MORPH_CLOSE, cross_selem)

# Display the results of the closing operation
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
axes[0, 0].imshow(image, cmap='gray')
axes[0, 0].set_title('Original Image')
axes[0, 0].axis('off')

axes[0, 1].imshow(closed_rectangle, cmap='gray')
axes[0, 1].set_title('Closing with Rectangle SE')
axes[0, 1].axis('off')

axes[1, 0].imshow(closed_circle, cmap='gray')
axes[1, 0].set_title('Closing with Circle SE')
axes[1, 0].axis('off')

axes[1, 1].imshow(closed_cross, cmap='gray')
axes[1, 1].set_title('Closing with Cross SE')
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

## **4 Morphological Gradient:**
● Implement the morphological gradient operation using OpenCV and logic.

● Apply the morphological gradient operation on a grayscale image using different structuring elements such as rectangles, circles, and crosses.
● Visualize and analyze the differences in results obtained with different
structuring elements.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = '/image.webp'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
kernel_rect = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
kernel_circle = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
kernel_cross = cv2.getStructuringElement(cv2.MORPH_CROSS, (5, 5))

# Apply morphological gradient operation
gradient_rect = cv2.morphologyEx(image, cv2.MORPH_GRADIENT, kernel_rect)
gradient_circle = cv2.morphologyEx(image, cv2.MORPH_GRADIENT, kernel_circle)
gradient_cross = cv2.morphologyEx(image, cv2.MORPH_GRADIENT, kernel_cross)

# Visualize the results
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title('Morphological Gradient (Rectangle)')
plt.imshow(gradient_rect, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title('Morphological Gradient (Circle)')
plt.imshow(gradient_circle, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title('Morphological Gradient (Cross)')
plt.imshow(gradient_cross, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## **5 Top Hat Operation:**
● Implement the top hat operation using OpenCV and logic.

● Apply the top hat operation on a grayscale image using different structuring
elements such as rectangles, circles, and crosses.

● Examine the output images and discuss the effects of different structuring
elements on the results.

```python
image_path = '/image.webp'  # Adjust the path as necessary
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
kernel_rect = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
kernel_circle = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
kernel_cross = cv2.getStructuringElement(cv2.MORPH_CROSS, (5, 5))

# Apply Top Hat operation
top_hat_rect = cv2.morphologyEx(image, cv2.MORPH_TOPHAT, kernel_rect)
top_hat_circle = cv2.morphologyEx(image, cv2.MORPH_TOPHAT, kernel_circle)
top_hat_cross = cv2.morphologyEx(image, cv2.MORPH_TOPHAT, kernel_cross)

# Visualize the results
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title('Top Hat (Rectangle)')
plt.imshow(top_hat_rect, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title('Top Hat (Circle)')
plt.imshow(top_hat_circle, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title('Top Hat (Cross)')
plt.imshow(top_hat_cross, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
