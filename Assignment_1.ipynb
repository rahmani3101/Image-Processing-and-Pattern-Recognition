```python
import cv2
from google.colab.patches import cv2_imshow # Import the correct function to display images in Colab

image = cv2.imread('/content/sample_data/asad.jpeg')

b,g,r = cv2.split(image)

cv2_imshow(image) # Use cv2_imshow instead of cv2.imshow
cv2_imshow(b)
cv2_imshow(g)
cv2_imshow(r)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

b, g, r = cv2.split(image)


merged_bgr = cv2.merge([b, g, r])
merged_gbr = cv2.merge([g, b, r])


cv2_imshow(merged_bgr)
cv2_imshow(merged_gbr)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python


from PIL import Image

# Load the image
image_path = '/content/sample_data/asad.jpeg'  # Replace with the actual path
image = Image.open(image_path)

# Extract information
width, height = image.size
format = image.format
mode = image.mode

# Print the information
print("Image Size:", width, "x", height)
print("Image Format:", format)
print("Image Mode:", mode)
```

```python
image_path = '/content/sample_data/asad.jpeg'  # Replace with the actual path
image = Image.open(image_path)

# Rotate the image
rotated_90 = image.rotate(90)
rotated_60 = image.rotate(60)
rotated_30 = image.rotate(30)
rotated_150 = image.rotate(150)
rotated_180 = image.rotate(180)

# Display the rotated images
display(rotated_90)
display(rotated_60)
display(rotated_30)
display(rotated_150)
display(rotated_180)
```

```python
image = cv2.imread('/content/sample_data/asad.jpeg')
flipped_horizontally = cv2.flip(image, 1)
flipped_vertically = cv2.flip(image, 0)
cv2_imshow(image)
cv2_imshow(flipped_horizontally)
cv2_imshow(flipped_vertically)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from PIL import Image, ImageDraw, ImageFont

image_path = '/content/sample_data/asad.jpeg'
image = Image.open(image_path)

draw = ImageDraw.Draw(image)

text = "Hello,Asad It is Taj Mahal"
text_position = (10, 10)

draw.text(text_position, text, fill=(255, 255, 255))

display(image)
```

```python


image_path = '/content/sample_data/asad.jpeg'
image = Image.open(image_path)

draw = ImageDraw.Draw(image)

draw.line((50, 50, 200, 50), fill=(255, 0, 0), width=3)  # Red line
draw.line((100, 100, 100, 250), fill=(0, 255, 0), width=5)  # Green line
draw.line((150, 150, 300, 300), fill=(0, 0, 255), width=2)  # Blue line

display(image)
```

```python
image_path = '/content/sample_data/asad.jpeg'
image = Image.open(image_path)

draw = ImageDraw.Draw(image)

top_left = (50, 50)
bottom_right = (200, 150)

draw.rectangle([top_left, bottom_right], outline=(0, 0, 255), width=3)

display(image)
```

```python
image_path = '/content/sample_data/asad.jpeg'
image = Image.open(image_path)

draw = ImageDraw.Draw(image)

points = [(100, 100), (200, 50), (250, 150), (300, 100),
          (350, 200), (300, 300), (200, 250), (150, 350),
          (100, 200), (50, 150)]

draw.polygon(points, outline=(255, 0, 0), fill=(0, 255, 0))

display(image)
```
