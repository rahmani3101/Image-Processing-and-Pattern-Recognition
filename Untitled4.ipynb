```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

class FuzzyCMeans:
    def __init__(self, n_clusters=3, m=2.0, max_iter=100, tol=1e-4):
        """
        Fuzzy C-Means Clustering Implementation

        Parameters:
        -----------
        n_clusters : int, number of clusters
        m : float, fuzzy coefficient (fuzziness parameter)
        max_iter : int, maximum iterations
        tol : float, convergence tolerance
        """
        self.n_clusters = n_clusters
        self.m = m  # Fuzzy coefficient
        self.max_iter = max_iter
        self.tol = tol

        self.centroids = None
        self.membership = None

    def _initialize_membership(self, X):
        """
        Initialize membership matrix randomly
        """
        n_samples = X.shape[0]
        membership = np.random.dirichlet(np.ones(self.n_clusters), size=n_samples)
        return membership

    def _update_centroids(self, X, membership):
        """
        Update cluster centroids
        """
        # Compute centroids using fuzzy membership
        membership_m = membership ** self.m
        centroids = np.zeros((self.n_clusters, X.shape[1]))

        for c in range(self.n_clusters):
            numerator = np.sum(membership_m[:, c][:, np.newaxis] * X, axis=0)
            denominator = np.sum(membership_m[:, c])
            centroids[c] = numerator / denominator

        return centroids

    def _update_membership(self, X, centroids):
        """
        Update membership matrix
        """
        # Compute distances between points and centroids
        distances = cdist(X, centroids)

        # Avoid division by zero
        distances = np.fmax(distances, np.finfo(float).eps)

        # Compute membership
        membership = np.zeros((X.shape[0], self.n_clusters))
        for c in range(self.n_clusters):
            membership[:, c] = np.sum(
                (distances[:, c][:, np.newaxis] / distances) ** (2 / (self.m - 1)),
                axis=1
            ) ** -1

        # Normalize membership
        membership /= np.sum(membership, axis=1)[:, np.newaxis]

        return membership

    def fit(self, X):
        """
        Fit Fuzzy C-Means to the data
        """
        # Initialize membership matrix
        self.membership = self._initialize_membership(X)

        for _ in range(self.max_iter):
            # Store previous centroids for convergence check
            prev_centroids = self.centroids

            # Update centroids
            self.centroids = self._update_centroids(X, self.membership)

            # Update membership
            new_membership = self._update_membership(X, self.centroids)

            # Check convergence
            if prev_centroids is not None:
                if np.linalg.norm(self.centroids - prev_centroids) < self.tol:
                    break

            self.membership = new_membership

        return self

    def predict(self, X):
        """
        Predict cluster labels based on highest membership
        """
        return np.argmax(self._update_membership(X, self.centroids), axis=1)

def generate_activity_dataset():
    """
    Generate synthetic Human Activity Recognition dataset
    """
    # Set random seed for reproducibility
    np.random.seed(42)

    # Generate data for different activities
    activities = {
        'Walking': {
            'acc_mean': [0.5, -0.3, 0.7],
            'gyro_mean': [0.1, 0.2, -0.1],
            'noise': 0.2
        },
        'Running': {
            'acc_mean': [1.2, -0.8, 1.5],
            'gyro_mean': [0.5, 0.4, -0.6],
            'noise': 0.3
        },
        'Standing': {
            'acc_mean': [0.1, 9.8, 0.1],
            'gyro_mean': [0.01, 0.02, 0.03],
            'noise': 0.1
        },
        'Sitting': {
            'acc_mean': [-1.5, 0.1, -0.8],
            'gyro_mean': [-0.3, -0.1, -0.2],
            'noise': 0.2
        }
    }

    # Collect data
    data = []

    # Generate samples for each activity
    # Get a list of activity names for indexing
    activity_names = list(activities.keys())
    for activity, params in activities.items():
        # Generate 100 samples for each activity
        for _ in range(100):
            # Generate accelerometer and gyroscope readings
            acc_data = np.array(params['acc_mean']) + \
                       np.random.normal(0, params['noise'], 3)
            gyro_data = np.array(params['gyro_mean']) + \
                        np.random.normal(0, params['noise'], 3)

            # Combine data
            sample = np.concatenate([acc_data, gyro_data])
            # Use activity_names.index to get the index
            data.append(np.append(sample, activity_names.index(activity)))

    # Create DataFrame
    columns = [
        'Acc_X', 'Acc_Y', 'Acc_Z',
        'Gyro_X', 'Gyro_Y', 'Gyro_Z',
        'Activity_Label'
    ]
    df = pd.DataFrame(data, columns=columns)

    return df

def perform_clustering_analysis(X, activity_labels):
    """
    Perform clustering analysis using Fuzzy C-Means and K-Means
    """
    # Different numbers of clusters to experiment with
    cluster_ranges = [3, 4, 5]

    plt.figure(figsize=(15, 10))

    for i, n_clusters in enumerate(cluster_ranges, 1):
        # Fuzzy C-Means
        fcm = FuzzyCMeans(n_clusters=n_clusters)
        fcm.fit(X)
        fcm_labels = fcm.predict(X)

        # K-Means
        kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)
        kmeans_labels = kmeans.fit_predict(X)

        # Visualization of first two features
        plt.subplot(2, 3, i)
        scatter = plt.scatter(X[:, 0], X[:, 1], c=fcm_labels,
                               cmap='viridis', alpha=0.7)
        plt.title(f'Fuzzy C-Means ({n_clusters} Clusters)')
        plt.xlabel('Accelerometer X')
        plt.ylabel('Accelerometer Y')
        plt.colorbar(scatter)

        # Membership intensity visualization
        plt.subplot(2, 3, i+3)
        membership_intensity = np.max(fcm.membership, axis=1)
        scatter = plt.scatter(X[:, 0], X[:, 1],
                               c=membership_intensity,
                               cmap='plasma', alpha=0.7)
        plt.title(f'Membership Intensity ({n_clusters} Clusters)')
        plt.xlabel('Accelerometer X')
        plt.ylabel('Accelerometer Y')
        plt.colorbar(scatter)

        # Print clustering results
        print(f"\nClustering Results for {n_clusters} Clusters:")

        # Compute and print cluster distribution
        print("\nFuzzy C-Means Cluster Distribution:")
        unique, counts = np.unique(fcm_labels, return_counts=True)
        for cluster, count in zip(unique, counts):
            print(f"Cluster {cluster}: {count} samples")

        print("\nK-Means Cluster Distribution:")
        unique, counts = np.unique(kmeans_labels, return_counts=True)
        for cluster, count in zip(unique, counts):
            print(f"Cluster {cluster}: {count} samples")

    plt.tight_layout()
    plt.savefig('clustering_comparison.png')
    plt.show()

def main():
    # Generate synthetic dataset
    df = generate_activity_dataset()

    # Separate features and labels
    X = df.iloc[:, :-1].values
    activity_labels = df.iloc[:, -1].values

    # Standardize features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Perform clustering analysis
    perform_clustering_analysis(X_scaled, activity_labels)

if __name__ == '__main__':
    main()
```
